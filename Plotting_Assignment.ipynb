```python
# YOUR CODE HERE
carbon = fuel_econ['co2']
range = np.arange(200, 800 , 100)
plt.hist(carbon, bins=range, edgecolor='red')
plt.xlabel('carbon dioxide')
plt.ylabel('rate')
plt.title('CARBON')
```

```python
plt.hist(fuel_econ.city, orientation='vertical', edgecolor='green')
plt.xlabel('city')
plt.ylabel('rate')
plt.title('Just city')
plt.show()
```

```python
grouped_data = fuel_econ.groupby('VClass')['city'].mean().reset_index()
VClass = grouped_data['VClass']
city = grouped_data['city']
plt.barh(VClass, city, color='blue')
plt.xlabel('City MPG (mean)')
plt.ylabel('Vehicle Class')
plt.title('Vehicle Class vs. City MPG (mean)')
plt.show()
```
